### A simple perceptron model with a single inner node and without activation function 

```python
#Declaring the inputs , weights and bias manually
inputs = [0.1,2.3,3.2]
weights = [3.3,2.0,3.1]
bias = 0.001
```

```python
#Formula : inputs * weights + bias

output = 0 
for i,j in zip(inputs,weights):
    output += i*j
finalOutput = output + bias
```

```python
#Final Output 
finalOutput
```

### A simple perceptron model with a single inner layer and without activation function 

```python
# this neural network contains 3 neurons with 4 inputs 
import numpy as np

inputs = [1,2,3,2.5]

weights = [[0.2,0.8,-0.5,1.0],
           [0.5,-0.91,0.26,-0.5],
           [-0.26,-0.27,0.17,0.87]]

biases = [2,3,0.5]
```

```python
# we will use the dot product method from numpy 
finalOutput = np.dot(weights,inputs) + biases
finalOutput
```
