#### This code demonstrates the implementation of Linear Regression using Neural Network approach

###### The 5-Step Model Life-Cycle


A model has a life-cycle, and this very simple knowledge provides the backbone for both modeling a dataset and understanding the tf.keras API.

The five steps in the life-cycle are as follows:

1. Define the model.
2. Compile the model.
3. Fit the model.
4. Evaluate the model.
5. Make predictions.

##### Reference :https://machinelearningmastery.com/tensorflow-tutorial-deep-learning-with-tf-keras/

```python
#importing relevant libraries 
import pandas as pd
from sklearn.model_selection import train_test_split
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
#loading the dataset
df = pd.read_csv('https://raw.githubusercontent.com/jbrownlee/Datasets/master/housing.csv',header=None)
```

```python
#declaring the variables and splitting the dataset into train and test 
X = df.values[:,:-1]
y = df.values[:,-1]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
print("Printing out the dimensions of the data after splitting :",X_train.shape,X_test.shape,y_train.shape,y_test.shape)
```

```python
#determining the number of input features
input_features = X_train.shape[1]
```

```python
#defining a model - Here we are building a multi-layer perceptron
model = Sequential()

#first hidden layer , the first argument for the Dense() is equal to the number of input features
model.add(Dense(13,activation='relu',kernel_initializer='he_normal',input_shape=(input_features,)))

#second hidden layer
model.add(Dense(8,activation='relu',kernel_initializer='he_normal'))

#output layer
model.add(Dense(1))
```

```python
#compiling the model 
model.compile(optimizer='adam',loss='mse')  #here we define Adam Optimizer and MSE as a loss function 
```

```python
#fitting the model
model.fit(X_train,y_train,batch_size=32,epochs=300,verbose=1)
```

```python
#evaluating the model
from numpy import sqrt
error = model.evaluate(X_test,y_test,verbose=1)
print("MSE : {} and RMSE : {} ".format(round(error,3),round(sqrt(error),3)))
```

```python
# making a prediction
import numpy as np
row = [0.00632,18.00,2.310,0,0.5380,6.5750,65.20,4.0900,1,296.0,15.30,396.90,4.98]
yhat = model.predict(np.array([row]))
print("Predicted value :",yhat)
```
